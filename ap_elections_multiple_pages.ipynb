```python
import requests
```

```python
url = "https://results.eci.gov.in/AcResultGenJune2024/statewiseS011.htm"
```

```python
response = requests.get(url)
```

```python
response
```

```python
response = response.content
```

```python
response
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(response,'html.parser')
```

```python
soup
```

```python
column_data = soup.find_all('td', align=lambda x: x in ['left', 'right'])
```

```python
column_data
```

```python
import pandas as pd
```

```python
bjp_seats_table_titles = ['Constituency', 'Constituency No', 'Leading Candidate', 'Trailing Candidate','Trailing Party', 'Margin','Round','Status']
```

```python
df = pd.DataFrame(columns = bjp_seats_table_titles)
```

```python
df
```

```python
extracted_data = []

# Assuming each set of data has a fixed structure of 8 columns (excluding leading_party)
for i in range(0, len(column_data), 8):
    if i + 7 < len(column_data):  # Check if there are enough elements for this row
        # Check if the current row contains unwanted pattern
        if 'Party Wise State Trends' in column_data[i+1].text.strip():
            continue  # Skip this row

        # Extract data from each group of 8 columns (excluding leading_party)
        constituency = column_data[i].text.strip()
        const_no = column_data[i+1].text.strip()
        leading_candidate = column_data[i+2].text.strip()
        trailing_candidate = column_data[i+3].text.strip()
        trailing_party = column_data[i+4].text.strip()
        margin = column_data[i+5].text.strip()
        round_info = column_data[i+6].text.strip()
        status = column_data[i+7].text.strip()

        # Append the extracted data as a tuple
        extracted_data.append((
            constituency,
            const_no,
            leading_candidate,
            trailing_candidate,
            trailing_party,
            margin,
            round_info,
            status
        ))
    else:
        print(f"Skipping incomplete row at index {i}")

# Print the extracted data for verification
for data in extracted_data:
    print(data)
```
